Problem 1

```python
import matplotlib.pyplot as plt

def rice_grains_by_day(n_days=100):
    grains_per_day = []
    total_grains = []
    total = 0
    
    for day in range(1, n_days + 1):
        grains_today = 2 ** (day - 1)
        total += grains_today
        grains_per_day.append(grains_today)
        total_grains.append(total)
    
    return grains_per_day, total_grains

# Get the data for 100 days
grains_per_day, total_grains = rice_grains_by_day()

# Plotting the graph for the number of grains per day and total grains
plt.figure(figsize=(10, 5))
plt.plot(range(1, 101), grains_per_day, label='Grains per Day', color='blue')
plt.plot(range(1, 101), total_grains, label='Total Grains', color='green')
plt.xlabel("Days")
plt.ylabel("Number of Grains")
plt.title("Number of Rice Grains over 100 Days")
plt.legend()
plt.show()

# Print the total number of grains on day 100
print("Total grains on day 100: {}".format(total_grains[-1]))
```

```python
def compute_sorori_shinzaemon(n_days=100):
    """
    A function that calculates the number of rice grains per day in the rice anecdote of Shinzaemon Sorori.
    
    Parameters
    -----------
    n_days: int
        Number of days to get rice (default: 100)
    
    Returns
    -------
    list_n_grains : list
        List of the number of rice grains received each day
    list_total_grains : list
        List of total rice grains received by each day
    """
    
    list_n_grains = []
    list_total_grains = []
    total = 0
    
    for day in range(1, n_days + 1):
        grains_today = 2 ** (day - 1)
        total += grains_today
        list_n_grains.append(grains_today)
        list_total_grains.append(total)
    
    return list_n_grains, list_total_grains

# Example: Calculate for 81 days
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=81)

# Plotting the graph for 81 days
plt.figure(figsize=(10, 5))
plt.plot(range(1, 82), list_n_grains, label='Grains per Day', color='blue')
plt.plot(range(1, 82), list_total_grains, label='Total Grains', color='green')
plt.xlabel("Days")
plt.ylabel("Number of Grains")
plt.title("Number of Rice Grains over 81 Days")
plt.legend()
plt.show()

# Print the total number of grains on day 81
print("Total grains on day 81: {}".format(list_total_grains[-1]))
```

```python
def days_of_life_from_rice(total_grains, num_people):
    """
    Calculate how many days a group of people can live with the given amount of rice.

    Parameters
    ----------
    total_grains: int
        The total number of grains of rice available.
    num_people: int
        The number of people to feed.

    Returns
    -------
    int
        The number of days the group can live with the rice.
    """
    
    # 1 grain of rice weighs 0.02 grams
    grams_per_grain = 0.02
    
    # 1 person needs 500 grams of rice per day
    rice_per_day_per_person = 500
    
    # Calculate total rice in grams
    total_rice_grams = total_grains * grams_per_grain
    
    # Calculate total rice needed for all people per day
    total_rice_needed_per_day = num_people * rice_per_day_per_person
    
    # Calculate how many days the group can live
    days_of_life = total_rice_grams // total_rice_needed_per_day
    
    return int(days_of_life)

# Example calculation: 10 people with the rice from 100 days
total_rice_100_days = total_grains[-1]
num_people = 10

days_to_live = days_of_life_from_rice(total_rice_100_days, num_people)
print(f"10 people can live for {days_to_live} days with the rice from 100 days.")
```
