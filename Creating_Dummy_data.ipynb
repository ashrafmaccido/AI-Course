Problem 1 - Creating Random Numbers

```python
import numpy as np
#create the variables
mean = [-3, 0]
cov = [[1, 0.8], [0.8, 1]]
np.random.seed(0)
rand_num = np.random.multivariate_normal(mean, cov, 500)
print(rand_num)
rand_num.shape
```

Problem 4 - Addition of Data

```python
import numpy as np
#create the variables
mean = [0, -3]
cov = [[1, 0.8], [0.8, 1]]
np.random.seed(0)
rand_num1 = np.random.multivariate_normal(mean, cov, 500)
print(rand_num)
rand_num.shape
```

```python
from matplotlib import pyplot as plt
%matplotlib inline

plt.title("Two-in-One Scatter Plot")
plt.xlabel("x1")
plt.ylabel("x2")

plt.scatter(rand_num[:,0], rand_num[:,1], s = 20, label = '0', marker = '*')
plt.scatter(rand_num1[:,0], rand_num1[:,1], s = 20, label = '1', marker = '*')
plt.legend()
plt.show()
```

Problem 5 - Data Concatenate

```python
# Using Data Concatenate
data_combine = np.concatenate((rand_num, rand_num1), axis=0)
print(data_combine)
data_combine.shape
```

```python
# Using vstack
data_combined = np.vstack((rand_num, rand_num1))
data_combined.shape
```

Problem 6 - Labeling

```python
# Create a (500, 3) ndarray
zeros = np.zeros(500)
ones = np.ones(500)
new_column = np.concatenate((zeros, ones))
new_ndarr = np.append(data_combined, new_column[:,None], 1)
print(new_ndarr)
new_ndarr.shape
```
